```python
import maabara as ma
import numpy as np
import scipy as sp
import sympy as sym
import matplotlib as mpl 
import matplotlib.pyplot as plt
import math
import matplotlib.gridspec as gridspec
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
mpl.rcParams['text.usetex']=True
mpl.rcParams['text.latex.unicode']=True
mpl.rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})


plt.figure()
ax=plt.subplot(111,aspect='equal')
x=np.linspace(0,20,1000)

ax.plot(x,10*np.exp(-x),'k--')
ax.plot(x,100*np.exp(-x),'k--')
ax.plot(x,np.exp(10.5)*np.exp(-x),'k--')
ax.plot(x,2*np.exp(12)*np.exp(-x),'k--')
ax.plot(x,3*np.exp(14.5)*np.exp(-x),'k--')

ax.plot(x,(2*(x+15))**(1/2.),label='$S=-15$')
ax.plot(x,(2*(x+10))**(1/2.),label='$S=-10$')
ax.plot(x,(2*(x+5))**(1/2.),label='$S=-5$')
ax.plot(x,(2*(x+0))**(1/2.),label='$S=0$')
x=np.linspace(5,20,1000)
ax.plot(x,(2*(x-5))**(1/2.),label='$S=5$')
x=np.linspace(10,20,1000)
ax.plot(x,(2*(x-10))**(1/2.),label='$S=10$')
x=np.linspace(15,20,1000)
ax.plot(x,(2*(x-15))**(1/2.),label='$S=15$')

ax.plot(x,10*np.exp(-x),'k--')
ax.plot(x,100*np.exp(-x),'k--')
ax.plot(x,np.exp(10.5)*np.exp(-x),'k--')
ax.plot(x,2*np.exp(12)*np.exp(-x),'k--')
ax.plot(x,3*np.exp(14.5)*np.exp(-x),'k--')

ax.arrow(10.5, 1,1/(2)**(1/2.),-1/(2)**(1/2.), head_width=0.1, head_length=0.1, fc='r', ec='r',zorder = 2)
ax.arrow(12, 2,1/(5)**(1/2.),-1/(5)**(1/2.)*2, head_width=0.1, head_length=0.1, fc='r', ec='r',zorder = 2)
ax.arrow(14.5, 3,1/(10)**(1/2.),-1/(10)**(1/2.)*3, head_width=0.1, head_length=0.1, fc='r', ec='r',zorder = 2)


ax.set_title('Phasenfronten und Ausbreitung der Lichtstrahlen')
ax.legend(numpoints=1, markerscale=1.5, fontsize=10,ncol=3,fancybox=True, shadow=True, loc=0)
ax.set_xticks(np.arange(0,20,1))
ax.set_yticks(np.arange(0,12,1))
ax.set_ylim(0,12)
ax.set_ylabel('y')
ax.set_xlabel('x')
ax.grid(True)

plt.savefig("plot.pdf",transparent=True,format="pdf",bbox_inches='tight')

plt.clf()
plt.figure()
ax2=plt.subplot(111,aspect='equal')
x=np.linspace(0,2,1000)
ax2.plot(x,(1+x**2)**(1/2.))
ax2.set_ylim(0.5,2.5)
ax2.set_xlabel('$\\frac{y}{h}$')
ax2.set_ylabel('$\\frac{n}{n_0}$')
ax2.grid(True)
plt.savefig("plot2.pdf",transparent=True,format="pdf",bbox_inches='tight')
```
